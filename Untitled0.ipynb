```python
#   Relatório de Pré-Decolagem — Aurora Siger

# 1. LEITURA DOS DADOS (telemetria simulada)
telemetria = {
    "temp_interna": 22,
    "temp_externa": -157,
    "integridade": 1,
    "energia": 87,
    "pressao_combustivel": 4.2,
    "pressao_oxigenio": 3.8,
    "mod_propulsao": 1,
    "mod_navegacao": 1,
    "mod_suporte_vida": 1,
    "mod_comunicacao": 0,
}

# 2. EXECUÇÃO DAS VERIFICAÇÕES
def verificar_telemetria(dados):
    alertas = []
    aprovado = True

    if not (18 <= dados["temp_interna"] <= 26):
        aprovado = False
        alertas.append("FALHA: Temperatura interna fora da faixa.")

    if dados["temp_externa"] >= -100:
        aprovado = False
        alertas.append("FALHA: Temperatura externa acima do limite.")

    if dados["integridade"] != 1:
        aprovado = False
        alertas.append("FALHA: Integridade estrutural comprometida.")

    if dados["energia"] <= 80:
        aprovado = False
        alertas.append("FALHA: Nível de energia insuficiente.")

    for tanque in ["pressao_combustivel", "pressao_oxigenio"]:
        if not (3.5 <= dados[tanque] <= 5.0):
            aprovado = False
            alertas.append(f"FALHA: Pressão do tanque {tanque} fora da faixa.")

    for modulo in ["mod_propulsao", "mod_navegacao", "mod_suporte_vida"]:
        if dados[modulo] != 1:
            aprovado = False
            alertas.append(f"FALHA: Módulo crítico {modulo} com defeito.")

    if dados["mod_comunicacao"] != 1:
        alertas.append("ALERTA: Módulo de comunicação com falha (não crítico).")

    return aprovado, alertas

# 3. RESULTADO FINAL IMPRESSO
print("===== RELATÓRIO DE PRÉ-DECOLAGEM — AURORA SIGER =====")
print()

aprovado, alertas = verificar_telemetria(telemetria)

if alertas:
    print("Ocorrências detectadas:")
    for msg in alertas:
        print(f"  → {msg}")
    print()

if aprovado:
    print("RESULTADO: ✓ PRONTO PARA DECOLAR")
else:
    print("RESULTADO: ✗ DECOLAGEM ABORTADA")
```
